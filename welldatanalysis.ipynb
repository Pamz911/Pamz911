```python
#TRAIN THE MODEL USING SUPPORT VECTOR MACHINE(SVM)
#one hot encoding for data
well_encoded = pd.get_dummies(total_attr_con)
print("Original DataFrame:")
print(well_encoded)
print("\nDataFrame after one-hot encoding:")
print(well_encoded)


```

```python
well_encoded.columns
```

```python
#Prediction of Swt
```

```python
label_encoder = LabelEncoder()
n_bins = 2
```

```python

#split the dataset into training and testing
X = well_encoded.drop(['Swt'], axis=1)
y = label_encoder.fit_transform(well_encoded['Swt'])
```

```python

from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)

from sklearn.svm import SVR

svr = SVR(kernel='rbf')
svr.fit(X_train, y_train)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 1. Fit the SVR model
svr_model = SVR()
svr_model.fit(X_train, y_train)

# 2. Predict on test data
y_pred = svr_model.predict(X_test)

# 3. Inverse transform predictions if necessary
# (If you applied any transformation before label encoding y, inverse transform the predictions)
# For example:
#y_pred = label_encoder.inverse_transform(y_pred)

# 4. Evaluate predictions
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # Compute RMSE by setting squared=False
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```
